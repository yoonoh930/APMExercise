# 7.1. Simulate a single predictor and a nonlinear relationship, such as a sin wave shown in Fig. 7.7, and investigate the relationship between the cost, Îµ, and kernel parameters for a support vector machine model:
<img src="img/fig7.7.png" style="height:800px">

```r
library(repr)
```

```r
set.seed(100)
x <- runif(100, min=2, max=10)
y <- sin(x) + rnorm(length(x))*.25
sinData <- data.frame(x=x, y=y)
options(repr.plot.width=7, repr.plot.height=4)
plot(x,y)
## Create a grid of x values to use for prediction
dataGrid <- data.frame(x=seq(2, 10, length=100))
```

### (a) Fit different models using a radial basis function and different values of the cost (the C parameter) and Îµ. Plot the fitted curve. For example:

```r
library(kernlab) #alpha is masked from 'package:ggplot2'
```

```r
rbfSVM <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar='automatic',
              C=0.05, epsilon=0.1)
rbfSVM1 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar='automatic',
              C=1, epsilon=0.1)
rbfSVM2 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar='automatic',
              C=20, epsilon=0.1)
modelPrediction <- predict(rbfSVM, newdata=dataGrid)
modelPrediction1 <- predict(rbfSVM1, newdata=dataGrid)
modelPrediction2 <- predict(rbfSVM2, newdata=dataGrid)

## This is a matrix with one column. We can plot the
## model predictions by adding points to the previous plot
options(repr.plot.width=7, repr.plot.height=4)
plot(x,y)
points(x=dataGrid$x, y=modelPrediction[,1],
      type='l', col='blue')
points(x=dataGrid$x, y=modelPrediction1[,1],
      type='l', col='red')
points(x=dataGrid$x, y=modelPrediction2[,1],
      type='l', col='black')
```

Support Vector Machine Regression minimize the following equation:

\begin{equation}
Cost \sum_{i=1}^N L_\epsilon(y_i - \hat{y}_i) + \sum_{j=1}^P\beta_j^2
\end{equation}

This we can say that as Cost increases, there is more viarance. It is clearly visible from the picture above

```r
rbfSVM3 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar='automatic',
              C=1, epsilon=0.001)
rbfSVM4 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar='automatic',
              C=1, epsilon=0.3)
rbfSVM5 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar='automatic',
              C=1, epsilon=1)
modelPrediction3 <- predict(rbfSVM3, newdata=dataGrid)
modelPrediction4 <- predict(rbfSVM4, newdata=dataGrid)
modelPrediction5 <- predict(rbfSVM5, newdata=dataGrid)

## This is a matrix with one column. We can plot the
## model predictions by adding points to the previous plot
options(repr.plot.width=7, repr.plot.height=4)
plot(x,y)
points(x=dataGrid$x, y=modelPrediction3[,1],
      type='l', col='blue')
points(x=dataGrid$x, y=modelPrediction4[,1],
      type='l', col='red')
points(x=dataGrid$x, y=modelPrediction5[,1],
      type='l', col='black')
```

We can also see that as epsilon increases, variance decreases. It is due to the nature of epsilon. It is used as a threshold to stop if the model is accurate enough. As epsilon increases, the standard gets loose thus having less variance

## (b) The Ï parameter can be adjusted using the kpar argument, such as kpar = list(sigma = 1). Try different values of Ï to understand how this parameter changes the model fit. How do the cost, Îµ, and Ï values affect the model?

```r
rbfSVM6 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar=list(sigma=0.1),
              C=1, epsilon=0.1)
rbfSVM7 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar=list(sigma=1),
              C=1, epsilon=0.1)
rbfSVM8 <- ksvm(x=x, y=y, data=sinData,
              kernel='rbfdot', kpar=list(sigma=100),
              C=1, epsilon=0.1)
modelPrediction6 <- predict(rbfSVM6, newdata=dataGrid)
modelPrediction7 <- predict(rbfSVM7, newdata=dataGrid)
modelPrediction8 <- predict(rbfSVM8, newdata=dataGrid)

## This is a matrix with one column. We can plot the
## model predictions by adding points to the previous plot
options(repr.plot.width=7, repr.plot.height=4)
plot(x,y)
points(x=dataGrid$x, y=modelPrediction6[,1],
      type='l', col='blue')
points(x=dataGrid$x, y=modelPrediction7[,1],
      type='l', col='red')
points(x=dataGrid$x, y=modelPrediction8[,1],
      type='l', col='black')
```

# Solution

```r
svmParam1 <- expand.grid(eps = c(0.01, 0.05, 0.1, 0.5), cost = 2^c(-2, 0, 2, 8))

for(i in 1:nrow(svmParam1)) {
    set.seed(131)
    rbfSVM <- ksvm(x = x, y = y, data=sinData,
                  kernel = "rbfdot", kpar = "automatic",
                  C = svmParam1$cost[i], epsilon = svmParam1$eps[i])
    
    tmp <- data.frame(x = dataGrid$x,
                     y = predict(rbfSVM, newdata = dataGrid),
                     eps = paste("epsilon:", format(svmParam1$eps)[i]),
                     costs = paste("cost:", format(svmParam1$cost)[i]))
    svmPred1 <- if(i == 1) tmp else rbind(tmp, svmPred1)
}
svmPred1$costs <- factor(svmPred1$costs, levels = rev(levels(svmPred1$costs)))
```

```r
set.seed(1016)
svmParam2 <- expand.grid(eps = c(0.01, 0.05, 0.1, 0.5), costs = 2^c(-2, 0, 2, 8),
                        sigma = as.vector(sigest(y~x, data = sinData, frac = 0.75)))
for (i in 1:nrow(svmParam2)) {
    rbfSVM <- ksvm(x = x, y = y, data=sinData,
                  kernel = "rbfdot",
                  kpar = list(sigma = svmParam2$sigma[i]),
                  C = svmParam2$costs[i],
                  epsilon = svmParam2$eps[i])
    tmp <- data.frame(x = dataGrid$x,
                      y = predict(rbfSVM, newdata = dataGrid),
                      eps = paste("epsilon:", format(svmParam2$eps)[i]),
                      costs = paste("cost:", format(svmParam2$costs)[i]),
                      sigma = paste("sigma:", format(svmParam2$sigma)[i])
                     )
    svmPred2 <- if(i == 1) tmp else rbind(tmp, svmPred2)
}
svmPred2$costs <- factor(svmPred2$costs, levels = rev(levels(svmPred2$costs)))
svmPred2$sigma <- factor(svmPred2$sigma, levels = rev(levels(svmPred2$sigma)))
```

# 7.2. Friedman (1991) introduced several benchmark data sets create by sim- ulation. One of these simulations used the following nonlinear equation to create data:

\begin{equation}
    y = 10sin(\pi x_1 x_2) + 20(x_3 - 0.5)^2 + 10x_4 + 5x_5 + N(0, \sigma^2)
\end{equation}

# where the x values are random variables uniformly distributed between [0, 1] (there are also 5 other non-informative variables also created in the simula- tion). The package mlbench contains a function called mlbench.friedman1 that simulates these data:

```r
library(caret)
library(mlbench)
```

```r
set.seed(200)
trainingData <- mlbench.friedman1(200, sd = 1)
## We convert the 'x' data from a matrix to a data frame
## One reason is that this will give the columns names.
trainingData$x <- data.frame(trainingData$x)
## Look at the data using
options(repr.plot.width=20, repr.plot.height=9)
featurePlot(trainingData$x, trainingData$y)
## or other methods
```

```r
## This creates a list with a vector 'y' and a matrix
## of predicitors 'x'. Also simulate a large test set to
## estimate the true error rate with good precision:
testData <- mlbench.friedman1(5000, sd = 1)
testData$x <- data.frame(testData$x)
```

```r
## Tune several models on theses data. For example
knnModel <- train(x = trainingData$x,
                  y = trainingData$y,
                 method = 'knn',
                 preProc = c("center", "scale"),
                 tuneLength = 10)
knnModel
```

```r
options(repr.plot.width=10, repr.plot.height=6)
plot(knnModel)
```

```r
knnPred <- predict(knnModel, newdata = testData$x)
## The function 'pstResample' can be used to get the test set
## performance values
postResample(pred = knnPred, obs = testData$y)
```

# Which models appear to give the best performance? Does MARS select the informative predictors (those named X1âX5)?

### Linear Regression

```r
set.seed(100)
linearModel <- train(x = trainingData$x,
                  y = trainingData$y,
                 method = 'lm',
                 preProc = c("center", "scale"))
lnPred <- predict(linearModel, newdata = testData$x)
pRSam <- list(linear = postResample(pred = lnPred, obs = testData$y))
print("Done with Linear Model")
ctrl <- trainControl(method = "cv")
print("Starting EARTH")
set.seed(100)
eModel <- train(x = trainingData$x,
               y = trainingData$y,
               method = 'earth',
               tuneGrid = expand.grid(degree = 1, nprune = 2:38),
               trControl = ctrl)
ePred <- predict(eModel, newdata = testData$x)
pRSam$earth <- postResample(pred = ePred, obs = testData$y)
print("Done with earth")
print("Starting SVM")
svmR <- train(x = trainingData$x,
             y = trainingData$y,
             method = "svmRadial",
             preProc = c("center", "scale"),
             tuneLength = 14,
             trControl = ctrl)
svmPred <- predict(svmR, newdata = testData$x)
pRSam$svmR <- postResample(pred = svmPred, obs = testData$y)
print("Done with svmR")
print("Starting knn")
pRSam$knn <- postResample(pred = knnPred, obs = testData$y)
print("Done")
```

# 7.3. For the Tecator data described in the last chapter, build SVM, neural network, MARS, and KNN models. Since neural networks are especially sen- sitive to highly correlated predictors, does pre-processing using PCA help the model?

```r
library(caret)
data(tecator)
library(reshape2)
```

```r
# Let's plot the graph
set.seed(42)
inSubset <- sample(1:dim(endpoints)[1], 10)

absorpSubset <- absorp[inSubset,]
#Protein subset
endpointsSubset <- endpoints[inSubset, 3]

specData <- as.data.frame(t(absorpSubset))
specData$x <- 1:nrow(specData)
specData2 <- melt(specData, id.vars = c("x"))
print(specData2)
```

```r
xyplot(value ~ x,
      data = specData2,
      groups = variable,
      type = c("l", "g"),
      panel = function(...) {
          panel.xyplot(...)
          panel.text(rep(101.5, nrow(absorpSubset)),
                    absorpSubset[, ncol(absorpSubset)],
                    paste(endpoints),
                    cex = 0.7)
      })
```

```r
## Dividing train/test set
set.seed(42)
idx_training <- createDataPartition(endpoints[, 3], p=3/4, list=FALSE)
Training_abrp <- as.data.frame(absorp[idx_training,])
Test_abrp <- as.data.frame(absorp[-idx_training,])
Training_pr <- endpoints[idx_training, 3]
Test_pr <- endpoints[-idx_training, 3]
```

```r
###### SVM without PCA
ctrl <- trainControl(method="cv")
svm_grid = expand.grid(sigma = c(0.01, 0.05, 0.1, 0.5),
                       C = 2^c(-2, -1, 0, 1, 2, 4))
set.seed(42)
svmR_No_PCA <- train(x = Training_abrp, y = Training_pr,
                    method = "svmRadial",
                    preProc = c("center", "scale"),
                    trControl = ctrl,
                    tuneGrid= svm_grid)
print("Done with No PCA")
##### SVM with PCA
set.seed(42)
svmR_PCA <- train(x = Training_abrp, y = Training_pr,
                    method = "svmRadial",
                    preProc = c("center", "scale", "pca"),
                    trControl = ctrl,
                    tuneGrid= svm_grid)
print("Done with PCA")
```

```r
pred_No_PCA <- predict(svmR_No_PCA, newdata = Test_abrp)
pred_PCA <- predict(svmR_PCA, newdata = Test_abrp)
postResample(pred = pred_No_PCA, obs = Test_pr)
postResample(pred = pred_PCA, obs = Test_pr)
```

So we can say that PCA actually affects model accuracy of SVMR in bad way. It is due to the the nature of PCA. PCA tries to reduce variance by grouping correlated variables. Then it produces linear combination of principle components. In the case above, it is decreasing the model performance

```r
plot(svmR_No_PCA,
    title = "No PCA")
plot(svmR_PCA,
    title = "PCA")
```
